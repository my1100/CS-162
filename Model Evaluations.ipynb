```python
from transformers import pipeline

detector = pipeline("text-classification", model="roberta-base-openai-detector", framework="pt")
```

```python
# replace this with your own folder path
folder_path = r"C:\Users\TP470EA-EC423W\Documents\CS 162\cs162-final-dev-main"
```

```python
file_paths = glob(os.path.join(folder_path, "*.jsonl"))

# load first 200 entries in each file
for file_path in file_paths:
    with open(file_path, "r", encoding="utf-8") as f:
        count = 0
        for line in f:
            if count >= 50:
                break

            entry = json.loads(line)

            texts.append(entry["human_text"])
            labels.append(0)

            texts.append(entry["machine_text"])
            labels.append(1)

            count += 1
```

```python
predictions = []

for text in texts:
    result = detector(text, truncation=True)[0]
    label = result["label"].lower()
    predicted = 1 if label == "fake" else 0
    predictions.append(predicted)
    print(predicted)
```

```python
trimmed_labels = labels[:len(predictions)]
accuracy = accuracy_score(trimmed_labels, predictions)
print(f"Accuracy: {accuracy:.4f}")
```

```python
report = classification_report(trimmed_labels, predictions, target_names=["Human", "AI"])

print(report)
```
